## Making Predictions

Now let's load the model and make some predictions:

```python
from predict import SentimentAnalyzer

# Load the trained model
analyzer = SentimentAnalyzer()

print("‚úÖ Model loaded successfully!")
```

### Try Some Examples

```python
# Example 1: Positive sentiment
text1 = "I absolutely loved this product! Best purchase ever!"
result1 = analyzer.predict(text1)

print(f"Text: {result1['text']}")
print(f"Sentiment: {result1['sentiment'].upper()}")
print(f"Confidence: {result1['confidence']:.2%}")
print(f"Probabilities: {result1['probabilities']}")
```

```python
# Example 2: Negative sentiment
text2 = "This was a terrible waste of money. Very disappointed."
result2 = analyzer.predict(text2)

print(f"Text: {result2['text']}")
print(f"Sentiment: {result2['sentiment'].upper()}")
print(f"Confidence: {result2['confidence']:.2%}")
print(f"Probabilities: {result2['probabilities']}")
```

## Interactive Testing

Try your own text:

## Batch Predictions

Analyze multiple texts at once:

```python
texts = [
    "Great product, highly recommend!",
    "Worst experience ever.",
    "It's okay, nothing special.",
    "Amazing quality and fast delivery!",
    "Poor quality, broke after one day."
]

results = analyzer.predict_batch(texts)

for i, result in enumerate(results, 1):
    print(f"\n{i}. {result['text']}")
    print(f"   Sentiment: {result['sentiment']} ({result['confidence']:.1%} confident)")
```

## Visualizing Results

Let's visualize the sentiment distribution:

```python
import matplotlib.pyplot as plt
import numpy as np

# Count sentiments
sentiments = [r['sentiment'] for r in results]
positive_count = sentiments.count('positive')
negative_count = sentiments.count('negative')

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(['Positive', 'Negative'], [positive_count, negative_count], 
        color=['#4CAF50', '#f44336'])
plt.title('Sentiment Distribution', fontsize=16, fontweight='bold')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Sentiment', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()

print(f"\nPositive: {positive_count} ({positive_count/len(results)*100:.1f}%)")
print(f"Negative: {negative_count} ({negative_count/len(results)*100:.1f}%)")
```
